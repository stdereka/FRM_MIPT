```python
from utils import *
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = {}
for i, name in enumerate(os.listdir('data')):
    data[i] = {}
    d = data[i]
    d['data'] = read_dat_file(os.path.join('data', name), ['lambda', 'trans'])
    params = get_params_from_file_name(name)
    d['temp'], d['width'], d['left_bound'], d['right_bound'] = params
```

```python
for i in range(len(data)):
    mes = data[i]
    X, Y = data[i]['data'].values.T
    plt.plot(X, Y)
    plt.xlim(mes['left_bound'], mes['right_bound'])
    plt.ylim(0, 100)
    plt.grid()
    plt.xlabel('$\lambda, nm$')
    plt.ylabel('$Transparency, %$')
    plt.show()
```
